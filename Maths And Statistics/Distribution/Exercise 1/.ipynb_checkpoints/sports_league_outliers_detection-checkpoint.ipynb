## Problem Statement


You are a data scientist working for a sports analytics company. The dataset, named **"sports_league_data.csv"** includes a variety of attributes for players in the league. The columns in the dataset are as follows:

- **player_id:** A unique identifier for each player.
- **age:** Age of the player.
- **position:** The position the player typically plays in (e.g., Forward, Midfielder, Defender, Goalkeeper).
- **team:** The team the player belongs to.
- **games_played:** The total number of games played by the player in the season.
- **player_rating:** An overall performance rating for the player for the season.

**Import Necessary Libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Task1

1. Import the data from the "sports_league_data.csv" file.
2. display the number of rows and columns. 
3. Display the first few rows of the dataset to get an overview.


```python
df = pd.read_csv('sports_league_data.csv')
print(df.shape)
df.head()
```

## Task2:

**Age Analysis:**
- Mean: Calculate the average age of players in the league.
- Standard Deviation Intervals:
    - Calculate the age range within one standard deviation of the mean.
    - Calculate the age range within two standard deviations of the mean.
    - Calculate the age range within three standard deviations of the mean.
  

```python
age_mean = round(df['age'].mean(),2)
age_std = round(df['age'].std(),2)

age_lower1 = round(age_mean - age_std,2)
age_upper1 = round(age_mean + age_std,2)

age_lower2 = round(age_mean - 2*age_std,2)
age_upper2 = round(age_mean + 2*age_std,2)

age_lower3 = round(age_mean - 3*age_std,2)
age_upper3 = round(age_mean + 3*age_std,2)

print("Player age range within one std. dev. of the mean: ",age_lower1,age_upper1)
print("Player age range within two std. dev. of the mean: ",age_lower2,age_upper2)
print("Player age range within three std. dev. of the mean: ",age_lower3,age_upper3)
```

## Task3:

**Player Rating Analysis:**
- Mean: Determine the average player rating across the league.
- Standard Deviation Intervals:
    - Calculate the player rating range within one standard deviation of the mean.
    - Calculate the player rating range within two standard deviations of the mean.
    - Calculate the player rating range within three standard deviations of the mean.

```python
rating_mean = round(df['player_rating'].mean(),2)
rating_std = round(df['player_rating'].std(),2)

rating_lower1 = round(rating_mean - rating_std,2)
rating_upper1 = round(rating_mean + rating_std,2)

rating_lower2 = round(rating_mean - 2*rating_std,2)
rating_upper2 = round(rating_mean + 2*rating_std,2)

rating_lower3 = round(rating_mean - 3*rating_std,2)
rating_upper3 = round(rating_mean + 3*rating_std,2)

print("Player rating range within one std. dev. of the mean: ",rating_lower1,rating_upper1)
print("Player rating range within two std. dev. of the mean: ",rating_lower2,rating_upper2)
print("Player rating range within three std. dev. of the mean: ",rating_lower3,rating_upper3)
```

## Task4:

Create histogram with Kernel Density Estimation (KDE) plots for both the 'age' and 'player_rating' columns from the dataset. While constructing these plots, ensure to visually represent the standard deviation intervals on the graph.


- **'age'**

```python
sns.histplot(df['age'],kde=True)
plt.axvline(age_mean, color='r', linestyle='--',linewidth=1)
plt.axvline(age_lower1, color='g', linestyle='--',linewidth=1)
plt.axvline(age_upper1, color='g', linestyle='--',linewidth=1)
plt.axvline(age_lower2, color='b', linestyle='--',linewidth=1)
plt.axvline(age_upper2, color='b', linestyle='--',linewidth=1)
plt.axvline(age_lower3, color='y', linestyle='--',linewidth=1)
plt.axvline(age_upper3, color='y', linestyle='--',linewidth=1)
plt.show()
```

- **'player_rating'**

```python
sns.histplot(df['player_rating'],kde=True)
plt.axvline(rating_mean, color='r', linestyle='--',linewidth=1)
plt.axvline(rating_lower1, color='g', linestyle='--',linewidth=1)
plt.axvline(rating_upper1, color='g', linestyle='--',linewidth=1)
plt.axvline(rating_lower2, color='b', linestyle='--',linewidth=1)
plt.axvline(rating_upper2, color='b', linestyle='--',linewidth=1)
plt.axvline(rating_lower3, color='y', linestyle='--',linewidth=1)
plt.axvline(rating_upper3, color='y', linestyle='--',linewidth=1)
plt.show()
```

## Task5

- Calculate the lower and upper boundaries for the 'Age' and 'Player_Rating' columns. 
- For each column, identify values that fall outside the calculated lower and upper boundaries. These values are considered outliers.

- **Display age outliers**

```python
df[(df['age']<age_lower3) | (df['age']>age_upper3)]
```

- **Display player_rating outliers**

```python
df[(df['player_rating']<rating_lower3) | (df['player_rating']>rating_upper3)]
```

## Task6

-  Remove outliers from both the 'age' and 'player_rating' columns and then visualize the adjusted data using histogram plots.


```python
df_no_outliers = df[((df['age']>age_lower3) & (df['age']<age_upper3)) & ((df['player_rating']>rating_lower3) & (df['player_rating']<rating_upper3))]
df_no_outliers
```

- **plot the histograms**

```python
sns.histplot(df_no_outliers['age'],kde=True)
```

```python
sns.histplot(df_no_outliers['player_rating'],kde=True)
```
